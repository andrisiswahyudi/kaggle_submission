```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('train.csv')
```

```python
df.head()
```

```python
df_k = df[['periode', 'kas_kantor']]
```

```python
df_k.head()
```

```python
df_k.drop(['periode'],1,inplace=True)
```

```python
predictionDays = 31
df_k['prediction'] = df_k[['kas_kantor']].shift(-predictionDays)
df_k.head()
```

```python
df_k.tail()
```

```python
x = np.array(df_k.drop(['prediction'],1))
x = x[:len(df_k)-predictionDays]
print(x)
```

```python
y = np.array(df_k['prediction'])
y = y[:-predictionDays]
print(y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = 0.2)
```

```python
predictionDays_array = np.array(df_k.drop(['prediction'], 1))[-predictionDays:]
print(predictionDays_array)
```

```python
from sklearn.svm import SVR
```

```python
svr_rbf = SVR(kernel='rbf', C=1e3, gamma=0.00001)
svr_rbf.fit(xtrain, ytrain)
```

```python
svr_rbf_confidence = svr_rbf.score(xtest, ytest)
print('SBR_RBF accuracy :', svr_rbf_confidence)
```

```python
svm_prediction = svr_rbf.predict(xtest)
print(svm_prediction)
print()
print(ytest)
```

```python
svm_prediction = svr_rbf.predict(predictionDays_array)
print(svm_prediction)
print()
print(df_k.tail(predictionDays))
```
